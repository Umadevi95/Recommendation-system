```python
import pandas as pd

# Load data
combined_data = pd.read_csv('combined_data.csv')

# Calculate the average rating for each movie
movie_avg_ratings = combined_data.groupby('movie_id')['rating'].mean().reset_index()
movie_avg_ratings.columns = ['movie_id', 'average_rating']

# Merge average ratings with the combined data
movies_with_avg_ratings = pd.merge(combined_data.drop_duplicates(subset='movie_id'), movie_avg_ratings, on='movie_id')

# List of genres
genres = ['Action', 'Adventure', 'Animation', "Children's", 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

# Dictionary to hold top movies for each genre
top_movies_by_genre = {}

# Get top 5 movies for each genre
for genre in genres:
    genre_movies = movies_with_avg_ratings[movies_with_avg_ratings[genre] == 1]
    top_n_movies = genre_movies.sort_values(by='average_rating', ascending=False).head(5)
    top_movies_by_genre[genre] = top_n_movies[['movie title', 'average_rating']]

# Convert the dictionary to a DataFrame for saving
top_movies_df = pd.concat(top_movies_by_genre, axis=0)

# Save the top movies by category to a CSV file
top_movies_df.to_csv('top_5_movies_by_category.csv', index=False)

print("Top 5 movies by category saved to top_5_movies_by_category.csv")
```
